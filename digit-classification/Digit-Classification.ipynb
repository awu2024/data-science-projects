# Handwritten Digit Classification Using SVD

### Form a matrix A for each digit, such that each row in A represents an image of that digit. (You will have 10 A's)

```python
# Import numpy and pandas
import numpy as np
import pandas as pd
```

```python
# read the trainInput.csv file
df_trainInput = pd.read_csv('svd_train/trainInput.csv', header=None)
df_trainInput
```

```python
# transpose trainInput
train_input_T = df_trainInput.T
train_input_T
```

```python
# read the trainOutput.csv file
df_trainOut = pd.read_csv('svd_train/trainOutput.csv', header=None)
df_trainOut
```

```python
# Convert trainOutput.csv file into an array
trainOut_array = df_trainOut.iloc[0]
trainOut_array
```

```python
# Make matrix A a dictionary to store 10 digits from 0-9
A = {1:[], 2:[], 3:[], 4:[], 5:[],
    6:[], 7:[], 8:[], 9:[], 0:[]}

# Make a for loop assign each row in trainInput to a digit in matrix A
for i in range(len(trainOut_array)):
    A[trainOut_array[i]].append(train_input_T.iloc[i])

# Convert each matrix A into a np.array to access the shape of the matrix more easily
# and print out the shape of each matrix A
for k in A.keys():
    A[k] = np.array(A[k])
    print(f'Digit {k}\'s matrix A shape: {A[k].shape}')
```

```python
# Printing out the numbers of rows for each matrix A 
# and check the total number of rows match the number of training digits
final = 0

print('Number of rows in each matrix A:')
for m in A.keys():
    print(f'    Digit {m}\'s matrix A rows: {len(A[m])}')
    final += len(A[m])
print(f'The total amount of rows across all matrices A is {final}')
```

Since there are a total of 1707 training digits and the total number of rows across all matrices A is 1707, we can conclude that all rows are correctly assigned to their respective matrices.

### Determine the singular value decomposition for each A. (Right singular vectors V, are an orthogonal basis in the image space of that digit. We will refer to the right singular vectors as "singular images.") You should get 10 sets of singular images, one for each digit.

```python
# Create a dictionary that stores 10 sets of singular digits

s_images = {}

# Use svd from np.linalg to get the right singular vectors for each digit
for d in A.keys():
    # Print out each matrix A shape to compare to SVD shapes later
    print(f'Digit {d}\'s matrix A shape: {A[d].shape}')
    u, s, vt = np.linalg.svd(A[d])
    # Print out the shape of left singular vectors (u), singular values (s),
    # and right singular vectors (vt)
    print(f'Digit {d}\'s singular value decomposition:')
    print(f'    Left singular vectors shape: {u.shape}')
    print(f'    Singular values length: {s.shape}')
    print(f'    Singular images (right singular vectors) shape: {vt.shape}')
    print()
    # Store each right singular vector to their respective digits
    s_images[d] = vt
```

Each matrix A has a shape of m x 256, where m is the number of rows (images) and 256 is the number of columns (pixel values). The singular images (right singular vectors) have a shape of 256 x 256, therefore, they have an orthogonal basis at the column space matrix A. We can conclude that the singular value decomposition has been correctly applied.

### Express test images as a linear combination of the first k=20 singular images of each digit. (This is a least square problem of the form Ax = b).

```python
# Read the testInput.csv file
df_testInput = pd.read_csv('svd_train/testInput.csv', header=None)
df_testInput
```

```python
# Transpose testInput
testInput_T = df_testInput.T
testInput_T
```

```python
# We need to express each test image as a linear combination by finding the coefficients (x)
# convert testInput_T into a np.array for easier access
test_image_array= np.array(testInput_T)

# Create a list that holds a dictionary of coefficients from digits 0-9 for each test image
x_list = []

# Loop through each test image
for row in range(len(test_image_array)):
    test_image = test_image_array[row]
    # Create a dictionary to store coefficients that would be appended later to x_list
    x_coef = {}
    # Loop through each digit to find the coefficient for each digit
    for d in s_images.keys():
        v_A = s_images[d][0:20].T
        x = np.linalg.pinv(v_A)@test_image
        x_coef[d] = x
    x_list.append(x_coef)
# Print out the first five elements of x_list
print(x_list[0:5])
```

### Compute the distance between test images and their least square approximations.

```python
# We need to find the least square approximation before we can compute the distance
# Create a list that holds the distance between test images and their least square approximations
distance_list = []

for row in range(len(test_image_array)):
    test_image = test_image_array[row]
    # Create a dictionary to store the distance
    distances_dict = {}
    for d in s_images.keys():
        v_A = s_images[d][0:20].T
        x = x_list[row][d]
        # Compute the least-square approximation with linear combination
        ls_approx = v_A @ x
        # Compute the distance between test images and ls_approx (||testImage - v_Ax||)
        distance = np.linalg.norm(test_image - ls_approx)
        distances_dict[d] = distance
    distance_list.append(distances_dict)
# Print out the first five elements of distance_list
print(distance_list[0:5])
```

### Classify each test image to be the digit corresponding to the smallest residual

```python
# We classify each unknown imaget to be the digit corresponding to the smallest error
# i.e. min||testImage - v_Ax||
# Create a list to store the digit corresponding to the smallest residual
digit_list = []
# Create a loop to check each distance in distance list for the smallest residual
for i in distance_list:
    temp_list = []
    for d in i.keys():
        temp_list.append(i[d])
    # Use the min function to find the smallest residual in the list
    min_digit = min(temp_list)
    # Finds the digit corresponding to the residual
    for c in i.keys():
        if min_digit == i[c]:
            digit_list.append(c)
            break;
# Print the list of predicted digits
print('List of predicted digits: ')
print(digit_list)
print(f'Predicted digit list length: {len(digit_list)}')
```

### Calculate the overall correct classification rate, as well as correct classification rate for each digit in a confusion matrix.

```python
# Load testOutput.csv
df_testOut = pd.read_csv('svd_train/testOutput.csv', header=None)
df_testOut
```

```python
# Overall classification rate = (Number of Correct Predictions/Total Number of Test Images) x 100
correct_pred = 0
testOut_list = df_testOut.iloc[0]
# Create a loop that checks how many digits from the testOutput list match our predicted list
for i in range(len(testOut_list)):
    if (testOut_list[i] == digit_list[i]):
        correct_pred += 1
# Use the formula above and print out the overall classification rate and correct predictions
print(f'Number of Correct Predictions: {correct_pred}')
class_rate = (correct_pred/len(testOut_list)) * 100
print(f'Overall correct classification rate: {class_rate}')
```

```python
# Import metrics from sklearn to access confusion matrix
from sklearn import metrics

confusion_m = metrics.confusion_matrix(testOut_list, digit_list)
# Print out the confusion matrix. Row (true labels) is read from top to bottom as 0-9 respectively
# and column (predicted labels) is read from left to right as 0-9 respectively
# The correct classification is along the diagonal for each digit 0-9 respectively
print('Confusion Matrix:')
print('Rows: True Labels (0-9) Columns: Predicted Labels (0-9)')
print(confusion_m)
```
